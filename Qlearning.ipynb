```python
#Algorithme d’apprentissage par renforcement où un agent apprend par essais-erreurs à choisir les meilleures actions 
#dans un environnement en maximisant une récompense cumulée. Son objectif est d’apprendre une politique optimale 
#sans modèle de l’environnement (model-free).
```

```python
# Ce code a pour objectif d’entraîner un agent avec Q-learning afin d’apprendre automatiquement la meilleure stratégie
#pour atteindre un état objectif (reward maximum) en se basant sur l’expérience, les essais et les erreurs, 
#et en construisant progressivement une Q-table qui représente la valeur de chaque action dans chaque état.
```

```python
import numpy as np  # importation de numpy pour les calculs numériques

# =========================================================
# 1. ENVIRONNEMENT (LE MONDE)
# =========================================================

n_states = 6  
# nombre total d’états (positions possibles de l’agent)

actions = [0, 1]  
# 0 = aller à gauche, 1 = aller à droite

rewards = np.array([0, 0, 0, 0, 0, 10])  
# récompenses : seule la dernière case donne un reward positif

# =========================================================
# 2. INITIALISATION DE LA Q-TABLE
# =========================================================

Q = np.zeros((n_states, len(actions)))  
# tableau Q initialisé à zéro (aucune connaissance au début)

# =========================================================
# 3. PARAMÈTRES DU MODÈLE
# =========================================================

alpha = 0.1  
# taux d’apprentissage (learning rate)

gamma = 0.9  
# facteur de réduction des récompenses futures

episodes = 100  
# nombre de fois où l’agent va apprendre

# =========================================================
# 4. APPRENTISSAGE (Q-LEARNING)
# =========================================================

for episode in range(episodes):  
# boucle sur toutes les simulations (expériences)

    state = 0  
    # l’agent commence toujours à l’état initial

    while state != n_states - 1:  
    # arrêter quand on atteint l’objectif final

        action = np.random.choice(actions)  
        # choix d’une action aléatoire (exploration)

        # =====================================================
        # transition vers le nouvel état
        # =====================================================

        if action == 1:  
            next_state = min(state + 1, n_states - 1)  
            # aller à droite sans dépasser la limite
        else:  
            next_state = max(state - 1, 0)  
            # aller à gauche sans dépasser 0

        reward = rewards[next_state]  
        # obtenir la récompense du nouvel état

        # =====================================================
        # MISE À JOUR DE LA Q-TABLE
        # =====================================================

        Q[state, action] = Q[state, action] + alpha * (
            reward + gamma * np.max(Q[next_state]) - Q[state, action]
        )  
        # formule principale de Q-learning :
        # correction de la valeur Q basée sur reward + futur optimal

        state = next_state  
        # déplacement vers le nouvel état

# =========================================================
# 5. RÉSULTAT FINAL
# =========================================================

print("Q-TABLE FINALE :")  
# affichage de la table d’apprentissage

print(Q)  
# valeurs finales apprises par l’agent
```
