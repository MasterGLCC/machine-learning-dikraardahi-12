```python
# ce code montre l'utilisation de la régression linéaire, polynomiale et multiple from scratch 

# ============================================
# 1. IMPORTATION DES BIBLIOTHÈQUES
# ============================================

import numpy as np  
# numpy : bibliothèque utilisée pour les calculs mathématiques 


import matplotlib.pyplot as plt  
# matplotlib : bibliothèque utilisée pour afficher les graphiques


# ============================================
# 2. DONNÉES
# ============================================

X_simple = np.array([
    230.1, 44.5, 17.2, 151.5, 180.8,
    8.7, 57.5, 120.2, 8.6, 199.8,
    66.1, 214.7, 23.8, 97.5, 204.1,
    195.4, 67.8, 281.4, 69.2, 147.3,
    218.4, 237.4, 13.2, 228.3, 62.3,
    262.9, 142.9, 240.1, 248.8, 70.6,
    74.7, 75.1, 123.1, 68.4, 135.2,
    255.4, 117.2, 76.4, 205.0, 182.6,
    11.7, 199.9, 7.3, 89.7, 191.1,
    175.1, 38.2, 39.5, 172.5, 85.3
])
# X_simple : variable d'entrée (feature)
# ici c'est le budget publicitaire TV
# chaque valeur représente une observation

y = np.array([
    22.1, 10.4, 12, 16.5, 17.9,
    7.2, 11.8, 13.2, 4.8, 15.6,
    12.6, 17.4, 9.2, 9.7, 19.0,
    22.4, 12.5, 24.4, 11.3, 14.6,
    18.0, 18.9, 5.3, 20.5, 9.7,
    17.0, 15.0, 20.9, 21.4, 11.0,
    11.2, 10.8, 13.9, 12.1, 14.3,
    19.8, 12.7, 11.5, 18.2, 17.6,
    6.1, 16.8, 3.9, 9.5, 17.1,
    15.9, 8.4, 7.8, 16.7, 10.2
])
# y : variable cible (target)
# ici c'est les ventes (Sales)


# ============================================
# 3. RÉGRESSION MULTIPLE (2 VARIABLES)
# ============================================

X_multiple = np.array([
    [230.1, 37.8],
    [44.5, 39.3],
    [17.2, 45.9],
    [151.5, 41.3],
    [180.8, 10.8],
    [8.7, 48.9],
    [57.5, 32.8],
    [120.2, 19.6],
    [8.6, 2.1],
    [199.8, 2.6],
    [66.1, 5.8],
    [214.7, 24.0],
    [23.8, 35.1],
    [97.5, 7.6],
    [204.1, 32.9],
    [195.4, 47.7],
    [67.8, 36.6],
    [281.4, 39.6],
    [69.2, 20.5],
    [147.3, 23.9],
    [218.4, 27.7],
    [237.4, 5.1],
    [13.2, 15.9],
    [228.3, 16.9],
    [62.3, 12.6],
    [262.9, 3.5],
    [142.9, 29.3],
    [240.1, 16.7],
    [248.8, 27.1],
    [70.6, 16.0],
    [74.7, 18.3],
    [75.1, 12.6],
    [123.1, 20.2],
    [68.4, 44.5],
    [135.2, 25.6],
    [255.4, 20.9],
    [117.2, 28.1],
    [76.4, 14.8],
    [205.0, 21.5],
    [182.6, 18.9],
    [11.7, 36.9],
    [199.9, 42.0],
    [7.3, 7.7],
    [89.7, 9.9],
    [191.1, 30.6],
    [175.1, 22.1],
    [38.2, 34.3],
    [39.5, 13.2],
    [172.5, 45.1],
    [85.3, 16.9]
])
# X_multiple : matrice des variables d'entrée
# colonne 1 = TV
# colonne 2 = Radio
# chaque ligne = une observation


# ============================================
# 4. FONCTION FROM SCRATCH (MOINDRES CARRÉS)
# ============================================

def fit_scratch(X_input, y_target):

    # X_input : données d'entrée
    # y_target : valeurs réelles (output)

    # ajout d'une colonne de 1 pour le biais (theta0)
    # cela permet d'apprendre l'intercept automatiquement
    X_b = np.c_[np.ones((len(X_input), 1)), X_input]

    # FORMULE DES MOINDRES CARRÉS :
    # theta = (Xᵀ · X)^(-1) · Xᵀ · y

    # Xᵀ = transpose de la matrice
    # dot = multiplication matricielle
    # inv = inverse matricielle

    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_target)

    return theta


# ============================================
# 5. RÉGRESSION LINÉAIRE SIMPLE
# ============================================

X_simple = X_simple.reshape(-1, 1)
# reshape : transforme en matrice colonne (obligatoire pour maths)

theta_simple = fit_scratch(X_simple, y)

print("===== Régression Linéaire Simple =====")
print("theta0 (bias / intercept) =", theta_simple[0])
# theta0 = biais (valeur quand X = 0)

print("theta1 (coefficient) =", theta_simple[1])
# theta1 = influence de TV sur les ventes


# ============================================
# 6. RÉGRESSION POLYNOMIALE
# ============================================

X_poly = np.c_[X_simple, X_simple**2]
# X² = ajout de non-linéarité

theta_poly = fit_scratch(X_poly, y)

print("\n===== Régression Polynomiale =====")
print("theta0 =", theta_poly[0])  # biais
print("theta1 =", theta_poly[1])  # X
print("theta2 =", theta_poly[2])  # X²


# ============================================
# 7. RÉGRESSION MULTIPLE
# ============================================

theta_multiple = fit_scratch(X_multiple, y)

print("\n===== Régression Multiple =====")
print("theta =", theta_multiple)
# theta = [biais, poids TV, poids Radio]


# ============================================
# 8. PRÉDICTIONS
# ============================================

y_pred_simple = theta_simple[0] + theta_simple[1] * X_simple
# modèle linéaire : y = θ0 + θ1X

y_pred_poly = (
    theta_poly[0]
    + theta_poly[1] * X_simple
    + theta_poly[2] * (X_simple**2)
)
# modèle polynomial : y = θ0 + θ1X + θ2X²

X_b_multiple = np.c_[np.ones((len(X_multiple), 1)), X_multiple]
# ajout du biais

y_pred_multiple = X_b_multiple.dot(theta_multiple)
# prédiction multiple


# ============================================
# 9. VISUALISATION
# ============================================

plt.figure(figsize=(8,5))
# création de la figure

plt.scatter(X_simple, y, label="Données réelles")
# points réels

plt.plot(X_simple, y_pred_simple, color="green" ,label="Linéaire")
# droite de régression

sorted_idx = np.argsort(X_simple[:,0])
# tri pour afficher une courbe propre

plt.plot(X_simple[sorted_idx], y_pred_poly[sorted_idx] , color="red", label="Polynomiale")
# courbe polynomiale

plt.title("Comparaison des modèles")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.legend()

plt.show()
```
